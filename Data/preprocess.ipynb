We already downsampled the Annotation dataset for $5%$, $10%$ and $1%$. We just need to propagate it to the questions.

```python
import pandas as pd
import json
```

```python
pct = 1

name_file = "annotations_sample_"+ str(pct)+ "pct" + ".json"
```

```python
annotation_path = "v2_Annotations_Val/" + name_file 
```

```python
question_base_path = "v2_Questions_Val/questions.json"
```

```python
with open(annotation_path, "r", encoding="utf-8") as f:
    annotation_json = json.load(f)
```

```python
annotation_pd = pd.DataFrame(annotation_json)
```

```python
question_id_tokeep = set(annotation_pd["question_id"].tolist())
```

```python
with open(question_base_path, "r", encoding="utf-8") as f:
    question_json = json.load(f)
```

```python
question_pd = pd.DataFrame(question_json["questions"])
```

```python
questions_tokeep_pd = question_pd[question_pd["question_id"].isin(question_id_tokeep)]
```

```python
questions_tokeep_pd.describe()
```

```python
save_path = "v2_Questions_Val/questions_sample_" + str(pct) + "pct" + ".json"

questions_tokeep_pd.to_json(save_path, orient="records")
```
